```r
# ============================================================
# 0. Dependencias
# ============================================================
if (!require(wooldridge)) install.packages("wooldridge")
if (!require(ggplot2))    install.packages("ggplot2")
if (!require(lmtest))     install.packages("lmtest")
if (!require(sandwich))   install.packages("sandwich")
if (!require(car))        install.packages("car")
if (!require(tseries))    install.packages("tseries")

library(wooldridge); library(ggplot2); library(lmtest)
library(sandwich);   library(car);     library(tseries)

# ============================================================
# 1. Datos
# ============================================================
data("beauty")

# Base original
beauty_original <- beauty

# Revisar datos nulos antes de limpiar
colSums(is.na(beauty_original))
sum(is.na(beauty_original))

# Eliminar observaciones con datos nulos
beauty <- na.omit(beauty_original)

# Comparar tamaño antes y después
dim(beauty_original)
dim(beauty)

# Estructura del dataset
print(names(beauty))
print(str(beauty))

# ============================================================
# 2. Estadísticas descriptivas
# ============================================================
summary(beauty[, c("wage", "looks", "educ", "exper", "female", "married", "black", "union")])

# Desviación estándar
sapply(beauty[, c("wage", "looks", "educ", "exper")], sd)

# ============================================================
# 3. Gráficos descriptivos
# ============================================================

# 3.1 Histograma de salario
ggplot(beauty, aes(x = wage)) +
  geom_histogram(binwidth = 1, fill = "steelblue", color = "black") +
  labs(title = "Distribución del Salario por Hora",
       x = "Salario por Hora (USD)", y = "Frecuencia") +
  theme_minimal()

# 3.2 Distribución de apariencia física (looks)
ggplot(beauty, aes(x = factor(looks))) +
  geom_bar(fill = "darkorange", color = "black") +
  labs(title = "Distribución del Puntaje de Apariencia Física",
       x = "Puntaje de Apariencia (1=muy poco atractivo, 5=muy atractivo)",
       y = "Frecuencia") +
  theme_minimal()

# 3.3 Salario vs. Apariencia
ggplot(beauty, aes(x = factor(looks), y = wage)) +
  geom_boxplot(fill = "lightblue") +
  labs(title = "Salario por Nivel de Apariencia Física",
       x = "Puntaje de Apariencia", y = "Salario por Hora (USD)") +
  theme_minimal()

# 3.4 Salario vs. Educación
ggplot(beauty, aes(x = educ, y = wage)) +
  geom_point(alpha = 0.4, color = "blue") +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Salario vs. Educación",
       x = "Años de Educación", y = "Salario por Hora (USD)") +
  theme_minimal()

# 3.5 Salario vs. Experiencia
ggplot(beauty, aes(x = exper, y = wage)) +
  geom_point(alpha = 0.4, color = "darkgreen") +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Salario vs. Experiencia",
       x = "Años de Experiencia", y = "Salario por Hora (USD)") +
  theme_minimal()

# ============================================================
# 4. Estimación MCO — Modelo Principal
# ============================================================
# Modelo: wage ~ looks + educ + exper + female + married + black + union
model <- lm(wage ~ looks + educ + exper + female + married + black + union,
            data = beauty)
summary(model)

# ============================================================
# 5. Verificación de Supuestos del MRLM
# ============================================================

# 5.1 Homocedasticidad — Breusch-Pagan
bptest(model)

# 5.2 Homocedasticidad — White
bptest(model,
       ~ I(looks^2) + I(educ^2) + I(exper^2) + looks*educ + female*married,
       data = beauty)

# 5.3 Corrección: errores estándar robustos HC1
coeftest(model, vcov = vcovHC(model, type = "HC1"))

# 5.4 Normalidad de los errores — Jarque-Bera
jarque.bera.test(residuals(model))

# 5.5 QQ-plot
qqnorm(residuals(model))
qqline(residuals(model), col = "red")

# 5.6 Gráfico: salario promedio según apariencia física
ggplot(beauty, aes(x = factor(looks), y = wage)) +
  stat_summary(fun = mean, geom = "bar", fill = "steelblue", color = "black") +
  labs(
    title = "Salario promedio según nivel de apariencia física",
    x = "Puntaje de apariencia física",
    y = "Salario promedio por hora"
  ) +
  theme_minimal()

# 5.7 Multicolinealidad — VIF
vif(model)

# ============================================================
# 6. Pruebas de Hipótesis
# ============================================================

# 6.1 Significancia conjunta (F-test) — incluido en summary(model)

# 6.2 H1: β_looks > 0  (prueba unilateral)
coefs   <- summary(model)$coefficients
t_looks <- coefs["looks", "t value"]
p_uni_looks <- pt(t_looks, df = model$df.residual, lower.tail = FALSE)
cat("p-valor unilateral (looks > 0):", round(p_uni_looks, 6), "\n")

# 6.3 H3: β_female < 0  (prueba unilateral)
t_female <- coefs["female", "t value"]
p_uni_female <- pt(t_female, df = model$df.residual, lower.tail = TRUE)
cat("p-valor unilateral (female < 0):", round(p_uni_female, 6), "\n")
```

1. Introducción
El presente informe aplica herramientas econométricas al conjunto de datos beauty del paquete wooldridge de R. El objetivo es identificar el efecto de la apariencia física y otras características individuales sobre el salario por hora de trabajadores estadounidenses, en línea con la literatura sobre discriminación y mercado laboral.
El dataset contiene 1,260 observaciones sobre trabajadores y fue recopilado por Hamermesh y Biddle (1994). La variable dependiente es wage (salario por hora en dólares), y las variables explicativas son looks (puntaje de apariencia), educ, exper, female, married, black y union.


## 2. Hipótesis de Investigación

Las hipótesis se derivan de la teoría del capital humano y la economía laboral:

**H1 — Apariencia física:** Una mayor puntuación en apariencia física aumenta el salario por hora.

$$H_0: \beta_{looks} = 0 \quad \text{vs.} \quad H_1: \beta_{looks} > 0$$

**H2 — Educación:** Más años de educación tienen un efecto positivo sobre el salario por hora.

$$H_0: \beta_{educ} = 0 \quad \text{vs.} \quad H_1: \beta_{educ} > 0$$

**H3 — Brecha de género:** Las mujeres perciben un salario por hora menor que los hombres, controlando por las demás variables.

$$H_0: \beta_{female} = 0 \quad \text{vs.} \quad H_1: \beta_{female} < 0$$

**H4 — Sindicato:** Pertenecer a un sindicato tiene un efecto positivo sobre el salario por hora.

$$H_0: \beta_{union} = 0 \quad \text{vs.} \quad H_1: \beta_{union} > 0$$

## 4. Especificación del Modelo

Se propone el siguiente **Modelo de Regresión Lineal Múltiple (MRLM)**:

$$wage_i = \beta_0 + \beta_1 \, looks_i + \beta_2 \, educ_i + \beta_3 \, exper_i + \beta_4 \, female_i + \beta_5 \, married_i + \beta_6 \, black_i + \beta_7 \, union_i + u_i$$

donde $u_i$ es el término de error que captura todos los factores no observados, y se asume que:

$$E(u_i \mid X_{i1}, X_{i2}, \dots, X_{i7}) = 0$$

Este supuesto de **exogeneidad** garantiza que los regresores no están correlacionados con el error, lo que permite la estimación insesgada de los coeficientes por Mínimos Cuadrados Ordinarios (MCO).

**Justificación de las variables:**
- $looks$: Variable de interés central. La teoría de la discriminación por apariencia predice un *beauty premium* en el mercado laboral.
- $educ$ y $exper$: Variables estándar de capital humano; se espera un efecto positivo sobre el salario.
- $female$: Captura la brecha salarial de género, esperada negativa según la literatura económica.
- $black$: Controla por posibles diferencias salariales por raza.
- $union$: El poder negociador sindical suele elevar los salarios.
- $married$: Puede reflejar mayor estabilidad laboral o efectos de selección.

- **`exper` ($\hat{\beta}_3$):** La experiencia laboral tiene un efecto positivo sobre

## 6. Verificación de Supuestos del MRLM

### 6.1 Exogeneidad

El MRLM exige que:

$$E(u_i \mid X_{i1}, X_{i2}, \dots, X_{i7}) = 0$$

Este supuesto implica que ninguna de las variables explicativas está correlacionada con
el término de error. En la práctica, se evalúa mediante argumentación teórica y análisis
de residuos.

La principal amenaza es el **sesgo por variable omitida**: la habilidad cognitiva no está
disponible en el dataset y podría estar correlacionada tanto con `looks` (si personas más
atractivas reciben mejor trato educativo o social) como con el salario. Esto podría sesgar
$\hat{\beta}_{looks}$ **hacia arriba**, sobreestimando el verdadero efecto de la apariencia
sobre el salario.

### 6.2 Homocedasticidad

El supuesto de homocedasticidad establece que la varianza del error es constante:

$$Var(u_i \mid X_{i1}, \dots, X_{i7}) = \sigma^2 \quad \forall \, i$$


```r
# Breusch-Pagan
bptest(model)

# White
bptest(model,
       ~ I(looks^2) + I(educ^2) + I(exper^2) + looks*educ + female*married,
       data = beauty)
       coeftest(model, vcov = vcovHC(model, type = "HC1"))
```

| Test | Estadístico | p-valor | Conclusión |
|---|---:|---:|---|
| Breusch-Pagan | 16.583 | 0.02029 | Se rechaza H0 al 5%; hay evidencia de heterocedasticidad. |
| White | 6.5781 | 0.6809 | No se rechaza H0 al 5%; no hay evidencia suficiente de heterocedasticidad. |

La hipótesis nula de ambos test es que existe homocedasticidad. Como el test de Breusch-Pagan presenta un p-valor menor a 0.05, se rechaza la hipótesis nula y se detecta evidencia de heterocedasticidad. En cambio, el test de White presenta un p-valor mayor a 0.05, por lo que no se rechaza la hipótesis nula. Debido a este resultado mixto, se decide reportar errores estándar robustos HC1 para realizar una inferencia más confiable.

```r
#Si p < 0.05, se rechaza homocedasticidad y se aplican errores estándar robustos de Huber-White:
coeftest(model, vcov = vcovHC(model, type = "HC1"))
```

## 7. Significancia Estadística y Prueba de Hipótesis

### 7.1 Significancia individual (t-tests)

Se evalúa si cada coeficiente es estadísticamente distinto de cero. Para cada regresor
se contrasta $H_0: \beta_k = 0$ vs. $H_1: \beta_k \neq 0$, reportando el p-valor
obtenido del output de `summary(model)` y concluyendo al nivel $\alpha = 0.05$.

### 7.2 Significancia conjunta (F-test)

Se prueba la hipótesis de que todos los coeficientes son simultáneamente iguales a cero:

$$H_0: \beta_1 = \beta_2 = \cdots = \beta_7 = 0$$

El estadístico F y su p-valor se obtienen directamente del output de `summary(model)`.
Si $p < 0.05$, se rechaza $H_0$ y el modelo en su conjunto es **estadísticamente significativo**.

### 7.3 Prueba unilateral — H1: $\beta_{looks} > 0$

Dado que la hipótesis es direccional, se aplica una **prueba t unilateral por la derecha**:

$$H_0: \beta_{looks} = 0 \quad \text{vs.} \quad H_1: \beta_{looks} > 0$$

Si $t_{looks} > 0$, el p-valor unilateral corresponde a la mitad del p-valor bilateral,
lo que fortalece el rechazo de $H_0$ y confirma que la apariencia física tiene un efecto
**positivo y significativo** sobre el salario por hora.

```r
coefs   <- summary(model)$coefficients
t_looks <- coefs["looks", "t value"]
p_uni_looks <- pt(t_looks, df = model$df.residual, lower.tail = FALSE)
cat("p-valor unilateral (looks > 0):", round(p_uni_looks, 6), "\n")
```

### 7.4 Prueba unilateral — H3: $\beta_{female} < 0$

Dado que la hipótesis plantea un efecto negativo del género femenino sobre el salario,
se aplica una **prueba t unilateral por la izquierda**:

$$H_0: \beta_{female} = 0 \quad \text{vs.} \quad H_1: \beta_{female} < 0$$

Si $t_{female} < 0$, el p-valor unilateral corresponde a la mitad del p-valor bilateral,
lo que fortalece el rechazo de $H_0$ y confirma la existencia de una **brecha salarial
de género** desfavorable para las mujeres, controlando por apariencia física, educación,
experiencia, estado civil, raza y afiliación sindical.

```r
t_female    <- coefs["female", "t value"]
p_uni_female <- pt(t_female, df = model$df.residual, lower.tail = TRUE)
cat("p-valor unilateral (female < 0):", round(p_uni_female, 6), "\n")
```

## 8. Conclusiones y Recomendaciones

### 8.1 Principales hallazgos

1. **H1 — Apariencia física (`looks`):** Se confirma la hipótesis. La apariencia física tiene un efecto positivo y estadísticamente significativo sobre el salario por hora. El coeficiente estimado es 0.40628 y su p-valor es menor a 0.05, lo que entrega evidencia a favor de la existencia de la hipótesis del *beauty premium* planteada por
   Hamermesh y Biddle (1994).

2. **H2 — Educación (`educ`):** Se confirma la hipótesis. La educación tiene un efecto positivo y significativo sobre el salario. Un año adicional de educación se asocia con un aumento promedio de 0.454 dólares por hora.

3. **H3 — Brecha de género (`female`):** Se confirma la hipótesis. El coeficiente de female es negativo y significativo, lo que indica que las mujeres ganan en promedio menos que los hombres, controlando por las demás características observables.

4. **H4 — Sindicato (`union`):** Se confirma la hipótesis. La afiliación sindical tiene un efecto positivo y significativo sobre el salario por hora.

En conjunto, los resultados sugieren que la apariencia física está positivamente asociada con el salario, incluso después de controlar por educación, experiencia, género, estado civil, raza y afiliación sindical. Sin embargo, los resultados deben interpretarse como una relación estadística y no necesariamente causal, debido a posibles variables omitidas como habilidad, ocupación, productividad o características del empleador.

### 8.2 Limitaciones del modelo

- La **asimetría positiva** del salario sugiere que un modelo log-lineal podría ser más
  apropiado:
  $$\ln(wage_i) = X_i\beta + u_i$$
  Esta transformación reduce la influencia de valores extremos y mejora la normalidad
  de los residuos.

- **Variables omitidas:** habilidad cognitiva, sector económico y tipo de contrato no
  están disponibles en el dataset, lo que puede generar sesgo en los coeficientes
  estimados, particularmente en $\hat{\beta}_{looks}$ y $\hat{\beta}_{educ}$.

- Si se detecta **heterocedasticidad**, los errores estándar OLS no son confiables para
  la inferencia y se debe reportar la versión corregida con errores robustos HC1.
