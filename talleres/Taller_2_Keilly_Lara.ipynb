**2.5. (15 puntos) Usando la regresión del salario en función de sales, roe y ros, se distinguen los
grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de
empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada
grupo son iguales**


```r
ceosal1_indus    <- subset(ceosal1, indus==1)
ceosal1_finance  <- subset(ceosal1, finance==1)
ceosal1_consprod <- subset(ceosal1, consprod==1)
ceosal1_utility  <- subset(ceosal1, utility==1)

modelo_pool     <- lm(salary ~ sales + roe + ros, data = ceosal1)
modelo_indus    <- lm(salary ~ sales + roe + ros, data = ceosal1_indus)
modelo_finance  <- lm(salary ~ sales + roe + ros, data = ceosal1_finance)
modelo_consprod <- lm(salary ~ sales + roe + ros, data = ceosal1_consprod)
modelo_utility  <- lm(salary ~ sales + roe + ros, data = ceosal1_utility)

SSR_r  <- sum(residuals(modelo_pool)^2)
SSR_ur <- sum(residuals(modelo_indus)^2)   +
          sum(residuals(modelo_finance)^2)  +
          sum(residuals(modelo_consprod)^2) +
          sum(residuals(modelo_utility)^2)

G   <- 4
k   <- 3
n   <- nrow(ceosal1)
gl1 <- (G - 1) * (k + 1)
gl2 <- n - G * (k + 1)

F_critico <- qf(0.95, gl1, gl2)
F_est     <- (SSR_r - SSR_ur) / SSR_ur * gl2 / gl1

cat("n:", n, "| gl1:", gl1, "| gl2:", gl2, "\n")
cat("SSR restringido:    ", round(SSR_r, 2), "\n")
cat("SSR no restringido: ", round(SSR_ur, 2), "\n")
cat("F estimado:         ", round(F_est, 4), "\n")
cat("F crítico (5%):     ", round(F_critico, 4), "\n")
if (F_est > F_critico) {"Se rechaza H0"} else {"No se rechaza H0"}
```

##2.5
Planteamiento de hipotesis

$H_0$: Los coeficientes son iguales para todos los sectores (el modelo es estable).  $H_1$: Al menos uno de los sectores tiene una estructura salarial distinta

Como el F estimado (1.3983) es menor que el F crítico (1.8026), el resultado es "No se rechaza H0".  

Al 5% de significancia, no existe evidencia estadística para afirmar que el sector de la empresa afecte la estructura salarial de los CEOs por lo que os modelos para cada grupo son estadísticamente iguales.
